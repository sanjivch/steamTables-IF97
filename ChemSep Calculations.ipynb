# ChemSep Calculations - Temperature correlations

```python
import numpy as np
import matplotlib.pyplot as plt
import math 
```

## Component - Air
### Ideal Gas Heat Capacity

```python
# Temperature in K
T = np.linspace(50,1500, 20)
```

```python
# Ideal gas heat capacity
A = 29562.29
B = -7.164949
C = 0.0216294
D = -0.0000139748
E = 2.89195E-09
```

```python
# Eq. 100 ChemSep
# y = A + Bt + Ct^2 + Dt^3 + Et^4
# Cp in J/kmol/K
Cp = A + B*T + C*T**2+ D*T**3 + E*T**4
```

```python
plt.plot(T,Cp, label='Ideal Gas Cp')
plt.xlabel('T (K)')
plt.ylabel('Cp (J/kgmol/K)')
plt.title('Ideal Gas Heat Capacity')
plt.grid()
```

### Vapor Pressure

```python
T = np.linspace(59.15, 132.5, 20)
```

```python
A = 14.794
B = -599.85
C = 1.0009
D = -3.9938E-07
E = 2
```

```python
# Eq 101 ChemSep
# y = exp(a + b/t + c.ln(t) + d.t^e)
# Pvap in Pa
Pvap = np.exp(A + B/T + C*np.log(T) + D*T**E)
```

```python
plt.plot(T, Pvap, label='Vapor Pressure')
plt.xlabel('T (K)')
plt.ylabel('Pvap (Pa)')
plt.title('Vapor Pressure')
plt.grid()
```

### Liquid Density

```python
T = np.linspace(59.15, 132.5, 20)
```

```python
A = 2.6731
B = 0.25637
C = 132.51
D = 0.26788
```

```python
# Eq 105 ChemSep
# y = a / b^[1 + (1 - t/c)^d]
# Lrho in kg/m3
Lrho = A/ (B**(1 + (1- T/C)**D))
```

```python
plt.plot(T, Lrho, label='Liquid Density')
plt.xlabel('T (K)')
plt.ylabel('Lrho (kg/m3)')
plt.title('Liquid Density')
plt.grid()
```

### Heat of Vaporisation

```python
T = np.linspace(59.15, 132.450, 20)
```

```python
A = 7385651
B = 0.276676
C = 0.211253
D = -0.836764
E = 0.722737
```

```python
# Eq 106 ChemSep 
# t=Tr   y = a(1 - t)^[b + c.t + d.t^2 + e.t^3]
# Hvap in J/kmol; Tr in K
Tr = 132.45 
Hvap = A*(1-T/Tr)**(B + C*T/Tr + D*(T/Tr)**2 + E*(T/Tr)**3)
```

```python
Hvap
```

```python
plt.plot(T, Hvap, label='Heat of Vaporisation')
plt.xlabel('T (K)')
plt.ylabel('Hvap (J/kmol)')
plt.title('Heat of Vaporisation')
plt.grid()
```
